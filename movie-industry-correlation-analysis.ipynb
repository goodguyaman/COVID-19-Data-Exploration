```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjust the configuration of the plots we will create

# Read the data

movies = '/kaggle/input/movies/movies.csv'

df = pd.read_csv(movies)
```

```python
# Let's look at the data
df.head()
```

```python
# Remove missing data
df = df.dropna()
```

```python
# Let's see if there's any missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
df.dtypes
```

```python
# Change the data type for budget

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct Year column
df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows',None)
```

```python
# Drop any duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
df.head()
```

```python
# Scatter plot eith budget vs gross revenue

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Buddget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Films')
plt.show()
```

```python
df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"pink"}, line_kws={"color":"green"})
```

```python
df.corr(method='pearson',numeric_only=True) #Pearson, Kendall, Spearman
```

```python
# High correlation between budget and gross, True
```

```python
correlation_matrix = df.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
